## Getting Started with the NLTK Toolkit

Now, we are going to move to using the NLTK Toolkit with Python. This is a popular package used by those in the digital humanities/computational linguistics, as it provides the ability to analyze, classify, tag linguistic data. The documentation guide is listed in the "Additional Helpful Links" section for those that want to examine the package more closely.

```python
#import packages
import nltk
from nltk.corpus import treebank
```

Next, we are going to create a variable named "sentence" as our input into the word tokenizer. You may choose your own sentence, or one of the following sentences:
* At eight o'clock on Thursday morning ... Arthur didn't feel very good. (PROVIDED BY NLTK PACKAGE)
* The curious cat pounced on a shiny red ball in the garden.
* She sipped her steaming coffee, lost in thought as the rain tapped on the window.
* The aroma of freshly baked bread wafted through the neighborhood, making everyone hungry.

_NOTE: I used ChatGPT to make these sentences, in case you were wondering why they were a bit weird üòÇ_

Once you have chosen your sentence, input the sentence into the middle of the quotes. Here, you may observer that we are using the tokenizer function, we will be using three sets of double quotes on both sides.


```python
sentence = """At eight o'clock on Thursday morning
... Arthur didn't feel very good."""
```

```python
#Tokenize the sentence. This function will separate the sentence into separate words.
tokens = nltk.word_tokenize(sentence)
tokens
```

```python
#Now, we will use the pos-tag function to categorize each word grammatically.
tagged = nltk.pos_tag(tokens)
print(tagged)
```

The abbreviations are in accordance to the Penn Treebank Project guide, which can be found here: https://www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html

Now, we can use the "chunk" function to categorize our words into segments.

```python
entities = nltk.chunk.ne_chunk(tagged)
print(entities)
```

## Conducting Sentiment Analysis

Sentiment analysis is a natural language processing (NLP) technique used to determine the overall emotional tone or sentiment expressed in a piece of text. It involves classifying a subtext as positive, negative, or neutral by assigning numerical scores to indicate the degree to which a text is positive or negative. We can use sentiment analysis

Sentiment analysis, also known as opinion mining, is a natural language processing (NLP) technique used to determine the emotional tone or sentiment expressed in a piece of text, such as a tweet, review, or news article. It involves analyzing the text to classify it as positive, negative, or neutral, or to assign a numerical score indicating the degree of positivity or negativity. Sentiment analysis has a wide range of applications, from social media monitoring to market research and customer feedback analysis.

Digital humanity students can use sentiment analysis to consider how the certain texts' affect constructs and/or reifies historical, social, and geo-political entanglements of power. For this workshop, we will be looking at Donald Trump's tweets related to immigration from 2020 and prior.

### Sentiment Analysis

Now that we have our tweets in a list, we will do some basic sentiment analysis on our tweets. We will need to import some additional packages.

```python
#import more packages
from tweepy import OAuthHandler
from textblob import TextBlob
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Define values
sentiment_values = [TextBlob(tweet) for tweet in tweets]

#IF YOU ARE USING TWEETS_SUBSTITUTE -> your code should be this instead
#sentiment_values = [TextBlob(tweet) for tweet in tweets_substitute]
```

```python
# Find the polarity values for each tweet -> .sentiment.polarity is the method that finds the polarity value in TextBlob
polarity_values = [[tweet.sentiment.polarity, str(tweet)] for tweet in sentiment_values]
```

### Polarity Values

The polarity value for sentiment analysis is a numerical value that represents the sentiment or emotion associated with a piece of text. The polarity value is a float number within the range of -1 to 1, where:
- A positive polarity value (close to 1) indicates a positive sentiment, meaning the text is positive, happy, or favorable.
- A negative polarity value (close to -1) indicates a negative sentiment, meaning the text is negative, unhappy, or unfavorable.
- A polarity value around 0 suggests a neutral sentiment, indicating that the text is neither positive nor negative, or it's a mix of bot

```python
#Print the first value to examine a bit closer
polarity_values[0]
```

```python
# Print all values
polarity_values[0:99]
```

Now, let's make a pandas dataframe that shows our tweets and their polarity.

```python
polarity_df = pd.DataFrame(polarity_values, columns=["polarity", "tweet"])

#display the first 20 entries (b/c we have over 1000 entries) -> if you used tweets_substitute, you should get 13 values
polarity_df.head(20)
```

### Let's create a wordcloud!

```python
# Create a Wordcloud from the tweets
tweets = ' '.join([text for text in tweets])
wordcloud = WordCloud(width=1400, height=1000, max_font_size=110).generate(tweets)

plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

# Let's make a pie chart!

```python
#First, let's count the number of positive, neutral, and negative polarities there are in our list of tweets based on 1, 0, and -1

#get all the polarity values
polarity_vals =polarity_df["polarity"]

positive = 0
neutral = 0
negative = 0

for polarity in polarity_vals:
    if polarity > 0:
        positive +=1
    if polarity < 0:
        negative += 1
    if polarity == 0:
        neutral += 1
```

```python
polarity_categories = ['Positive', 'Neutral', 'Negative']
polarity_counts = [positive, neutral, negative]

plt.pie(polarity_counts, labels=polarity_categories, autopct='%1.1f%%', startangle=140)
plt.title("Polarity Distribution")

plt.show()
```

# Let's make a bar chart!

```python
sentiment_categories = ['Positive', 'Neutral', 'Negative']
counts = [positive, negative, neutral]  # Replace with the actual counts for each category

plt.bar(sentiment_categories, counts, color=['green', 'blue', 'red'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Categories')
plt.ylabel('Counts')

plt.show()
```
